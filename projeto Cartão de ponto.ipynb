```python
class Horario:
    def __init__(self,hora,minuto,segundo):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo

    def __repr__(self):
        return f"{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}"

    def __add__(self, other):
        
        hora = self.hora + other.hora
        minuto = self.minuto + other.minuto
        segundo = self.segundo + other.segundo

        while segundo >= 60:
            minuto = minuto + 1
            segundo = segundo - 60

        while minuto >= 60:
            hora = hora + 1
            minuto = minuto - 60

        while hora >= 24:
            hora = hora - 24

        return Horario(hora, minuto, segundo)
    
    def __gt__(self, other):
        
        if self.hora > other.hora:
            
            return True
        
        elif self.hora == other.hora and self.minuto > other.minuto:
            
            return True
        
        elif self.hora == other.hora and self.minuto == other.minuto and self.segundo > other.segundo:
            
            return True
        
        else:
            
            return False




class sistemaDePonto:
    def __init__(self):
        self.horario = {}
        
        
    def interface (self):
        print('='*60)
        print('Sistema de Marcação de horarios'.center(60))
        print('='*60)


        print("\n1 - Marcar o horario de entrada no serviço")
        print('2 - Marcar o horario de saida para almoço')
        print('3 - Marcar o horario de entrada do almoço')
        print('4 - Marcar o horario de saida para casa')
        print("0 - Sair")
        opcao = int(input('Oque deseja fazer? '))

        while opcao != 0:
            if opcao == 1:
                self.chegada()

            elif opcao == 2:
                self.saidaAlmoco()

            elif opcao == 3:
                self.entradaPosAlmoco()

            elif opcao == 4:
                self.saidaNoite()

            else:
                print('opção invalida, digite uma opçao valida: ')
            opcao = int(input('Oque deseja fazer? '))
            

        
    def chegada(self):
        h,m,s = input('insira o horario de entrada no serviço, no formato hora minuto segundo: ').split()
        horarioEntrada = Horario(h,m,s)
        self.horario['entrada'] = horarioEntrada
    
    def saidaAlmoco(self):
        h,m,s = input('insira o horario de saida para o almoço, no formato hora minuto segundo: ').split()
        horarioSaidaAlmoço = Horario(h,m,s)
        self.horario['saidaAlmoço'] = horarioSaidaAlmoço
        
    def entradaPosAlmoco(self):
        horaDeAlmoço = Horario(1,0,0)
        h,m,s = input('insira o horario de saida para o almoço, no formato hora minuto segundo: ').split()
        horarioEntradaAlmoço = Horario(h,m,s)
        if horarioEntradaAlmoço < horaDeAlmoço or horarioEntradaAlmoço == horaDeAlmoço:
            print('Você não pode bater o ponto ainda, você não completou uma hora de almoço')
        else:
            self.horario['entradaAlmoço'] = horarioEntradaAlmoço
    
    def saidaNoite(self):
        horarioMaximo = Horario(10,0,0)
        h,m,s = input('insira o horario de saida do serviço, no formato hora minuto segundo: ').split()
        horarioSaida = Horario(h,m,s)
        self.horario['saida'] = horarioSaida
        horarioTotal = (self.horario.saidaAlmoço-self.horario.entrada) + (self.horario.saida-self.horario.entradaAlmoço)
        
        if horarioTotal >= horarioMaximo:
            print('Você trabalhou mais de 10h hoje. Por favor, pense na sua familia, amigos e seus pets e não segundo sobrecarrege, eles te amam')
        else:
            print ('Muito obrigado pelo trabalho, bom descanso')
    
    
sistema = sistemaDePonto()
sistema.interface()
```
